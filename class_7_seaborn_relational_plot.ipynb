```python
""" Class 7. Seaborn Relational Plots

Introduction to Seaborn:
- A python library for making statistical graphs
- It is built on top of matplotlib and easy-to-go with numpy and pandas
- It helps to explore and understand the data

""" Class 7. Seaborn Relational Plots

Introduction to Seaborn:
- A python library for making statistical graphs
- It is built on top of matplotlib and easy-to-go with numpy and pandas
- It helps to explore and understand the data

import seaborn as sns

sns.set_style(style='darkgrid')
sns.__version__
```

```python
""" Data Science
- business data
- data analysis
  - data cleaning
  - data preprocessing
  - data visualization
- Exploratory data analysis: Data analysis with visualization

Relational plots in seaborn:
   Establishes the relationship between two variables/features
   For example, a tabular data containing room size, no of bed rooms, owner, rent
   An example contains the values of each features. [1400, 2, 15K] (Feature vector)

If we have a ML model that predicts room rent (target variable),
   room_size, rent has significant relationship
   owner, rent has no significant relationship
"""

tips = sns.load_dataset('tips')

tips.head()
```

```python
tips.info()
```

```python
tips['time'].value_counts()
```

```python
tips['day'].value_counts()
```

```python
tips['smoker'].value_counts()
```

```python
tips['sex'].value_counts()
```

```python
""" Ask and Verify
Ask: Is there any relationship between total bill and tip?
Verify: Has a positive relation i.e. if total bill increases we may expect that the tip
will also increase.

Relational plots (kind='scatter'):
  1. Scatterplot (scatter)
  2. Line plot (line)
"""

sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    kind='scatter',
)
```

```python
"""
Ask 2: Is there any relation of tip based on days?
Verify 2: People like to tip more on saturday/sunday
"""

sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='day',
)
```

```python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='day',
    row='day'
)
```

```python
"""
Ask 3. Is there any relation of tip based on time?
Verify 3: Yes, people like to tip more on Dinner.
"""

sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex',
    col='time',
    style='smoker'
)
```

```python
""" Line plot """

fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
fmri.info()
```

```python
fmri['region'].value_counts()
```

```python
fmri['event'].value_counts()
```

```python
""" Region: Frontal, Parietal
Event: stim, cue
"""

sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    kind='line',
    height=3,
    col='region',
    row='event',
    hue='region',
)
```

```python
sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    kind='line',
    hue='event',
    col='subject',
    col_wrap=4,
    linewidth=2.5,
)
```
